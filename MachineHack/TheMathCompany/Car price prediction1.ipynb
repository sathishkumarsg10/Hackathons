```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_squared_log_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostRegressor
import math
cv = KFold()
LR = LinearRegression()
RF = RandomForestRegressor(n_estimators = 250, min_samples_leaf = 15)
GB = GradientBoostingRegressor()
AB = AdaBoostRegressor()
XGB = xgb.XGBRegressor()
LGB = lgb.LGBMRegressor()
CB = CatBoostRegressor()
```

```python
def model_eval(X, Y, model = 'LR'):
    xTrain, xTest, yTrain, yTest = train_test_split(X, Y, test_size = 0.2, random_state = 21)
    xTrain1, xVal, yTrain1, yVal = train_test_split(xTrain, yTrain, test_size = 0.2, random_state = 21) 
    if model == 'LR':
        LR.fit(xTrain1, yTrain1)
        pred_val = LR.predict(xVal)
        pred = LR.predict(xTest)
    elif model == 'RF':
        RF.fit(xTrain1, yTrain1)
        pred_val = RF.predict(xVal)
        pred = RF.predict(xTest)
    elif model == 'GB':
        GB.fit(xTrain1, yTrain1)
        pred_val = GB.predict(xVal)
        pred = GB.predict(xTest)
    elif model == 'AB':
        AB.fit(xTrain1, yTrain1)
        pred_val = AB.predict(xVal)
        pred = AB.predict(xTest)
    elif model == 'XGB':
        XGB.fit(xTrain1, yTrain1)
        pred_val = XGB.predict(xVal)
        pred = XGB.predict(xTest)
    elif model == 'LGB':
        LGB.fit(xTrain1, yTrain1)
        pred_val = LGB.predict(xVal)
        pred = LGB.predict(xTest)
    elif model == 'CB':
        CB.fit(xTrain1, yTrain1)
        pred_val = CB.predict(xVal)
        pred = CB.predict(xTest)
    print('Validation set: \n','MAE: ',mean_absolute_error(yVal, pred_val), '\nRMSE: ',math.sqrt(mean_squared_error(yVal, pred_val)),
                  '\nR2: ',r2_score(yVal, pred_val), '\nRMSLE: ',math.sqrt(mean_squared_log_error(yVal, pred_val)))
    print('MAE: ',mean_absolute_error(yTest, pred), '\nRMSE: ',math.sqrt(mean_squared_error(yTest, pred)),
                  '\nR2: ',r2_score(yTest, pred), '\nRMSLE: ',math.sqrt(mean_squared_log_error(yTest, pred)))
    return pred
```

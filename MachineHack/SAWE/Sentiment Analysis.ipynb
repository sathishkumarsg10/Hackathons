# Data Pre-processing

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
nltk.download('punkt')
```

```python
import re
import emoji
import string
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import PorterStemmer
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
porter = PorterStemmer()
emojis = emoji.UNICODE_EMOJI
```

```python
def remove_URL(text):
    URL = re.compile(r'https?://\S+|www/./S+')
    return URL.sub(r'',text)
def remove_handles(text):
    handle = re.compile(r'@\S+')
    return handle.sub(r'',text)
def remove_HTML(text):
    HTML = re.compile(r'<.*?>')
    return HTML.sub(r'', text)
def remove_punctuation(text):
    return ''.join([ch for ch in text if ch not in exclude])
def remove_stopwords(text):
    return ' '.join([x for x in word_tokenize(text) if x not in stop])
def remove_emojis(text):
    emojilist = [x for x in text if x in emojis]
    return ''.join([ch for ch in text if ch not in emojilist])
def stemming(text):
    return ' '.join([porter.stem(x) for x in word_tokenize(text)])
def remove_spaces(text):
    text = text.strip()
    text = text.split()
    return ' '.join(text)
```

```python
def data_preprocessing(text):
    text = text.apply(lambda x: x.lower())
    text = text.apply(lambda x: remove_URL(x))
    text = text.apply(lambda x: remove_handles(x))
    text = text.apply(lambda x: remove_HTML(x))
    text = text.apply(lambda x: remove_emojis(x))
    text = text.apply(lambda x: remove_punctuation(x))
    text = text.apply(lambda x: remove_spaces(x))
    text = text.apply(lambda x: remove_stopwords(x))
    text = text.apply(lambda x: stemming(x))
    return text
```

```python
train_text = data_preprocessing(train_text)
```

```python
test_text = data_preprocessing(test_text)
```

```python
data = data_preprocessing(data)
```

# Importing Glove Vectors 

```python
word_to_vec_map = {}
with open('/Users/sathishkumars/Documents/Anaconda-Python/glove/glove.6B.50d.txt') as file:
    for line in file:
        values = line.strip().split()
        curr_word = values[0]
        word_to_vec_map[curr_word] = np.array(values[1:], dtype = np.float64)
len(word_to_vec_map)
```

```python
dim = 50
```

```python
embedding_matrix = np.zeros((len(index) + 1, dim))
for word, i in index.items():
    temp = word_to_vec_map.get(word)
    if temp is not None:
        embedding_matrix[i] = temp
```

# 1) Bidirectional LSTM

```python
from keras.models import Sequential 
from keras.layers import Embedding, Bidirectional, LSTM, Dropout, Dense, GlobalMaxPooling1D
```

```python
embed_layer = Embedding(len(index) +1, dim, input_length = max_seq_length, weights = [embedding_matrix])
```

```python
model = Sequential()
model.add(embed_layer)
model.add(Bidirectional(LSTM(dim, return_sequences = True, dropout = 0.1, recurrent_dropout = 0.1)))
model.add(GlobalMaxPooling1D())
model.add(Dense(50, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(3, activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])
```

```python
history = model.fit(xTrain, yTrain, epochs = 2, batch_size = 128, validation_data = (xTest, yTest))
```
